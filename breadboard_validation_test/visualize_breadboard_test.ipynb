```python
import sys
import numpy as np
import matplotlib
import matplotlib.pylab as plt
%matplotlib inline
```

```python
file = open("data.txt", 'r')
lines = file.readlines()
print("File open ok")
```

```python
filtervals = []
emissionvals = []
```

```python
count = 0
for line in lines:
    count += 1
    if(line.strip().startswith("FIL")):
        filtervals.append(int(line.strip().replace("FIL ", ""))/1023)
    if(line.strip().startswith("EMM")):
        emissionvals.append(int(line.strip().replace("EMM ", ""))/1023)
        
print("Array population ok")
```

```python
# COLE RADETICH
#simple trick to widen the graph, gives a better view
config = plt.figure()
config.set_figwidth(20)
config.set_figheight(10)
#plot all_as vs all_accs
plt.plot(filtervals)
plt.title('Breadboard Test Emission')
plt.ylabel('Light Value (out of 1)')
plt.xlabel('Time')
plt.show()
```

```python
# COLE RADETICH
#simple trick to widen the graph, gives a better view
config = plt.figure()
config.set_figwidth(20)
config.set_figheight(10)
#plot all_as vs all_accs
plt.plot(emissionvals)
plt.title('Breadboard Test Filter')
plt.ylabel('Light Value (out of 1)')
plt.xlabel('Time')
plt.show()
```
